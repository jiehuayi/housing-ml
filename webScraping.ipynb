```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, WebDriverException
import pandas as pd
import time

path = "chromedriver.exe"
WEBSITES = {
    'langley': 'https://www.zoocasa.com/langley-bc-real-estate/sold?page=',
    'surrey': 'https://www.zoocasa.com/surrey-bc-real-estate/sold?page=',
    'burnaby': 'https://www.zoocasa.com/burnaby-bc-real-estate/sold?page=',
    'richmond': 'https://www.zoocasa.com/richmond-bc-real-estate/sold?page=',
    'vancouver': 'https://www.zoocasa.com/vancouver-bc-real-estate/sold?page=',
    'abbotsford': 'https://www.zoocasa.com/abbotsford-bc-real-estate/sold?page=',
    'coquitlam': 'https://www.zoocasa.com/coquitlam-bc-real-estate/sold?page=',
    'mapleRidge': 'https://www.zoocasa.com/maple-ridge-bc-real-estate/sold?page=',
    'delta': 'https://www.zoocasa.com/delta-bc-real-estate/sold?page=',
    'pittMeadows': 'https://www.zoocasa.com/pitt-meadows-bc-real-estate/sold?page=',
    'newWestminster': 'https://www.zoocasa.com/new-westminster-bc-real-estate/sold?page=',
    'whiteRock': 'https://www.zoocasa.com/white-rock-bc-real-estate/sold?page=',
}

# for scraping in ascending order
# WEBSITES = {
#     'langley-asc': 'https://www.zoocasa.com/langley-bc-real-estate/sold?sort=date&page=',
#     'surrey-asc': 'https://www.zoocasa.com/surrey-bc-real-estate/sold?sort=date&page=',
#     'burnaby-asc': 'https://www.zoocasa.com/burnaby-bc-real-estate/sold?sort=date&page=',
#     'richmond-asc': 'https://www.zoocasa.com/richmond-bc-real-estate/sold?sort=date&page=',
#     'vancouver-asc': 'https://www.zoocasa.com/vancouver-bc-real-estate/sold?sort=date&page=',
#     'abbotsford-asc': 'https://www.zoocasa.com/abbotsford-bc-real-estate/sold?sort=date&page=',
#     'coquitlam-asc': 'https://www.zoocasa.com/coquitlam-bc-real-estate/sold?sort=date&page=',
#     'mapleRidge-asc': 'https://www.zoocasa.com/maple-ridge-bc-real-estate/sold?sort=date&page=',
#     'delta-asc': 'https://www.zoocasa.com/delta-bc-real-estate/sold?sort=date&page=',
#     'pittMeadows-asc': 'https://www.zoocasa.com/pitt-meadows-bc-real-estate/sold?sort=date&page=',
#     'newWestminster-asc': 'https://www.zoocasa.com/new-westminster-bc-real-estate/sold?sort=date&page=',
#     'whiteRock-asc': 'https://www.zoocasa.com/white-rock-bc-real-estate/sold?sort=date&page=',
# }

email = "" # Add your email
password = "" # Add your password

XPATHS = {
    'login': './/a[contains(@data-testid,"loginHeader")]',
    'email': './/input[contains(@data-testid,"loginRegistrationEmailTextField")]',
    'continue': './/button[contains(@data-testid,"loginRegistrationContinueButton")]',
    'password': './/input[contains(@data-testid,"passwordLoginModal")]',
    'signin': './/button[contains(@data-testid,"signInButton")]',
    'view': './/button[contains(@data-testid,"overlaySignInBtn")]',
    'tnc': './/button[contains(@data-testid,"termsOfUseAcceptButton_31017")]',
    'close': '#listing > div.style_bar__mddLI > div > svg > g > g > path',
    'listingCard': './/div[contains(@class,"style_wrapper__Z8WDF")]',
    'address': './/h1[contains(@data-testid,"streetAddress")]',
    'beds': './/span[contains(@data-testid,"listingBedIcon")]',
    'bath': './/span[contains(@data-testid,"listingBathIcon")]',
    'area': './/span[contains(@data-testid,"listingDimensionsIcon")]',
    'parking': './/span[contains(@data-testid,"listingCarIcon")]',
    'maintenance': './/span[contains(@data-testid,"MaintenanceFeesKeyFacts")]',
    'type': './/span[contains(@data-testid,"TypeKeyFacts")]',
    'taxes': './/span[contains(@data-testid,"TaxesKeyFacts")]',
    'built': './/span[contains(@data-testid,"Approx.AgeKeyFacts")]',
    'mls': './/span[contains(@data-testid,"MLSÂ®NumberKeyFacts")]',
    'levels': './/span[contains(@data-testid,"LevelsKeyFacts")]',
    'garage': './/span[contains(@data-testid,"GarageKeyFacts")]',
    'lotSize': './/span[contains(@data-testid,"LotSizeKeyFacts")]',
    'pricePerSqFt': './/span[contains(@data-testid,"Avg.PricePerSqftKeyFacts")]',
    'soldDate': './/div[contains(@data-testid,"priceHistoryRowSoldDate")]',
    'soldPrice': './/div[contains(@data-testid,"priceHistoryRowSoldPrice")]',
    'listingDate': './/div[contains(@data-testid,"priceHistoryRowDate")]',
    'listingPrice': './/div[contains(@data-testid,"priceHistoryRowListPrice")]'
}

def create_driver():
    try:
        service = Service(executable_path=path)
        options = webdriver.ChromeOptions()
        driver = webdriver.Chrome(service=service, options=options)
        return driver
    except WebDriverException as e:
        print(f"Failed to create WebDriver: {e}")
        return None

def login(driver):
    try:
        loginBtn = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, XPATHS['login'])))
        loginBtn.click()

        emailBtn = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, XPATHS['email'])))
        emailBtn.send_keys(email)

        continueBtn = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, XPATHS['continue'])))
        continueBtn.click()

        passwordBtn = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, XPATHS['password'])))
        passwordBtn.send_keys(password)

        signInBtn = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, XPATHS['signin'])))
        signInBtn.click()

        view = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, XPATHS['view'])))
        view.click()

        tnc = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, XPATHS['tnc'])))
        tnc.click()

        close = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, XPATHS['close'])))
        close.click()

    except TimeoutException:
        print("Login failed due to timeout.")
        driver.quit()

def appendData(driver, list, text):
    try:
        list.append(driver.find_element(By.XPATH, XPATHS[text]).text)
    except NoSuchElementException:
        list.append(None)

def getData(site, start_page, end_page):
    driver = create_driver()
    if driver is None:
        return pd.DataFrame()

    driver.set_window_size(1920, 1080)
    driver.get(site)
    login(driver)
    print('done logging in')

    listingsData = pd.DataFrame(columns=[
        'address', 'beds', 'bath', 'area', 'maintenance', 'parking', 'taxes', 
        'built', 'mls', 'levels', 'garage', 'lotSize', 
        'pricePerSqFt', 'soldDate', 'soldPrice', 'listingDate', 'listingPrice'
    ])
        
    for page_num in range(start_page, end_page):
        driver.get(site + str(page_num))
        time.sleep(1)
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)
            
        try:
            listingCards = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, XPATHS['listingCard'])))
        except TimeoutException:
            print(f"Failed to load listing cards on page {page_num}")
            continue
            
        for card in listingCards:
            try:
                card.click()
                time.sleep(0.5)
                        
                data = {key: [] for key in listingsData.columns}
                for key in data.keys():
                    appendData(driver, data[key], key)
                
                new_data = pd.DataFrame(data)
                listingsData = pd.concat([listingsData, new_data], ignore_index=True)

                close = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, XPATHS['close'])))
                close.click()
                time.sleep(0.5)

            except (NoSuchElementException, TimeoutException, WebDriverException) as e:
                print(f"An error occurred with a listing card on page {page_num}: {e}")
                driver.get(site + str(page_num))
                time.sleep(0.5)
                continue
            print(f"Data loaded for page {page_num}")
    driver.quit()
    return listingsData

```

```python
def scrape_all_pages(city):
    final_data = pd.DataFrame()
    for i in range(1,51,1):  # Incrementing by 3 to read 3 pages at a time
        listings_data = getData(WEBSITES[city], i, min(i + 1, 51))
        final_data = pd.concat([final_data, listings_data], ignore_index=True)
        final_data.to_csv(f'data/ree{city}-listings.csv')
        # final_data.to_csv(f'data/predict/{city}-active-listings.csv')
```
